***Гіпотеза 2.***

Розбивка по місяцях.

Дослідимо кількість покупок по місяцях.

```python
# відберемо лише місяць
def get_month(Date):
    return int(Date.split('/')[0])
```

```python
# створимо колонку по місяцях
df['month'] = df.Date.apply(get_month)
df.head()
```

```python
# кількість покупок по місяцях
df['month'].value_counts()
```

```python
# підрахуємо прибуток по місяцях
monthly_profit = df.groupby(df['month'])['gross income'].sum()
display(monthly_profit)
```

Аналіз даних показав, що кількість покупок та прибуток значно зменшуються у лютому, відразу після святкового періоду в січні. Це свідчить про сезонний спад активності покупців. Для підвищення продажів у лютому необхідно організувати ефективні акції, розпродажі та нові пропозиції. Оптимальним буде розпочати рекламну кампанію вже в січні, щоб забезпечити успішний результат у лютому.

***Гіпотеза 3.***

Ефективність статуту учасника клубу.

Дослідимо, як відрізняються покупки учасників клубу магазину та просто покупців.

```python
# кількість покупок згідно з типом покупця
df['Customer type'].value_counts()
```

```python
# прибуток по категоріях за типами покупців
invoices_by_member = pd.pivot_table(df, values='gross income', index='Customer type', columns='Product line', aggfunc='sum')
display(invoices_by_member)
```

```python
# середня вартість чеку
invoices_by_member_price = pd.pivot_table(df, values='Unit price', index='Customer type', columns='Product line', aggfunc='mean')
display(invoices_by_member_price)
```

Згідно з даними, різниця в кількості покупок, середньому чеку та прибутку між членами клубу магазину і простими покупцями є незначною, часто не перевищує 10%, іноді прибуток від простих покупців навіть вищий. Це свідчить про те, що поточна політика клубної програми є неефективною.

Можливі причини:

*1. Недостатня привабливість клубної програми*

Члени клубу не отримують значних переваг, які б спонукали їх робити більше покупок. Це може означати, що програма лояльності або клубна програма не є достатньо привабливою або вигідною, щоб мотивувати членів купувати частіше.

*2. Схожі рівні покупок*

Можливо, що загальна привабливість магазину або продуктів є основним фактором, і рівень покупок не залежить від статусу члена клубу. Це може свідчити про те, що всі покупці, незалежно від їх статусу, мають подібні вподобання і купують у магазині з однаковою частотою.

*3. Ефективність маркетингових акцій*

Можливо, маркетингові акції і пропозиції однаково впливають на обидві групи покупців. Це може вказувати на те, що акції ефективні як для членів клубу, так і для простих покупців, і не створюють додаткової мотивації для більш частих покупок членів клубу.
