````python
from math import pi # Змінила * на pi, додала порожній рядок після імпорту

def calc_area(r): # Змінено назву функції з підкресленням та нижнім регістом, перенесла return на новий рядок
    return pi * r ** 2 # Чотири пробіли на початку, а також пробіли наколо операторів

def calc_circumference(r): # Змінено назву функції з підкресленням та нижнім регістом, перенесла return на новий рядок
    return 2 * pi * r # Чотири пробіли на початку, а також наколо операторів

radius = 5 # Пробіли навколо оператора
print("Area:", calc_area(radius), "Circumference:", calc_circumference(radius)) # Пробіли після коми
````

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})
````

````python
memory_before = df.memory_usage(index=False).sum()
memory_before
````

````python
df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')
````

````python
memory_after = df.memory_usage(index=False).sum()
memory_after
````

````python
memory_before - memory_after
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````

````python
input_list = [75, 80, 85, 90, 95]
````

````python
def calculate_variance(num_list):
    """Calculate the variance of a list of numbers."""
    mean = sum(num_list) / len(num_list)
    mean_squares = sum(n ** 2 for n in num_list) / len(num_list)

    return mean_squares - mean ** 2
````
